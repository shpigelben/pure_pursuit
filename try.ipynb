```python
def wrapper(gen_func):
    def wrapped(*args, **kwargs):
        print(f"running {gen_func.__name__}")
        return gen_func(*args, **kwargs)
    return wrapped

@wrapper
def square(x):
    return x**2

square(2)
```

```python
from dataclasses import dataclass
import numpy as np
from matplotlib.patches import Polygon

@dataclass
class RegularPolygon:
    n_sides: int
    radius: float = 1.0

    @property
    def vertices(self):
        angles = np.linspace(0, 2 * np.pi, self.n_sides, endpoint=False)
        x = self.radius * np.cos(angles)
        y = self.radius * np.sin(angles)
        return np.column_stack((x, y))
```

```python
p = RegularPolygon(n_sides=5, radius=2.0)

fig, ax = plt.subplots()
polygon_patch = Polygon(p.vertices, closed=True, fill=None, edgecolor='r')
ax.add_patch(polygon_patch)
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
plt.gca().set_aspect('equal', adjustable='box')

```
